Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("data.csv",encoding="latin")
print(df.head())
```

Handeling the Missing Values

```python
df["Description"] = df["Description"].fillna("Unknown")
df["CustomerID"] = df["CustomerID"].fillna(0)

df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]
```

```python
print(df.info())
```

Reviewing all core features

```python
print("===== WEEK 2 REVIEW =====")

# Feature 1
print("\n✅ Feature 1: Total Amount Calculation")
print(df[["Quantity", "UnitPrice", "TotalAmount"]].head())

# Feature 2
print("\n✅ Feature 2: Top 10 Best Selling Products")
top_products = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
print(top_products)

# Feature 3
print("\n✅ Feature 3: Top 10 Countries by Total Sales")
country_sales = df.groupby("Country")["TotalAmount"].sum().sort_values(ascending=False).head(10)
print(country_sales)

print("\n🎉 All core features are working correctly!")
```
