Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("data.csv",encoding="latin")
print(df.head())
```

Handeling the Missing Values

```python
df["Description"] = df["Description"].fillna("Unknown")
```

```python
df["Description"].isnull().sum()
```

```python
df["CustomerID"] = df["CustomerID"].fillna("Unknown")
```

```python
df["CustomerID"].isnull().sum()
```

Removing the Duplicate Values

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

Fixing Incorrect Datatypes

```python
df.dtypes
```

```python
df["InvoiceDate"].head(10)
```

Converting InvoiceDate from str into datetime


```python
df["InvoiceDate"] = pd.to_datetime(
    df["InvoiceDate"],
    format="%m-%d-%Y %H:%M",
    errors="coerce"
)
```

```python
df["InvoiceDate"].isna().sum()
```
