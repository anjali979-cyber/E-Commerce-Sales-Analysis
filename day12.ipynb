Importing required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("data.csv",encoding="latin")
print(df.head())
```

Handeling the Missing Values

```python
df["CustomerID"] = df["CustomerID"].fillna("Unknown")
```

```python
df["Description"] = df["Description"].fillna("Unknown")
```

```python
print(df.info)
```

Writting Basic Tests


```python
# Feature1 : Create TotalAmount feature
df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]

# Display the first five rows
df[["Quantity", "UnitPrice", "TotalAmount"]].head()
```

```python
# Test Feature 1: Total Amount Calculation

print("Feature 1 Test")
print(df["TotalAmount"].iloc[0])
```

```python
# Feature 2: Top 10 Best Selling Products

top_products = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)

print("Top 10 Best Selling Products:")
print(top_products)
```

```python
# Test Feature 2: Top 10 Best Selling Products

top_products = (
    df.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_products)
```

```python
# Core Feature 3: Top 10 Countries by Total Sales

country_sales = (
    df.groupby("Country")["TotalAmount"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(country_sales)
```

```python
# Test Feature 3: Top 10 Countries by Total Sales

country_sales = (
    df.groupby("Country")["TotalAmount"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(country_sales)
```

Running all tests at once

```python
print("===== Running All Feature Tests =====")

# Feature 1 Test
print("\nFeature 1: TotalAmount")
print(df["TotalAmount"].head())

# Feature 2 Test
print("\nFeature 2: Top 10 Best Selling Products")
top_products = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
print(top_products)

# Feature 3 Test
print("\nFeature 3: Top 10 Countries by Total Sales")
country_sales = df.groupby("Country")["TotalAmount"].sum().sort_values(ascending=False).head(10)
print(country_sales)

print("\n✅ All three feature tests completed successfully.")
```

Testing the complete Flow

```python
print("===== Complete Flow Test (10 Runs) =====")

for i in range(10):
    print(f"\nTest Run {i+1}")

    # Feature 1
    total_amount = df["TotalAmount"].iloc[i]
    print("Feature 1 - Total Amount:", total_amount)

    # Feature 2
    top_products = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
    print("Feature 2 - Top Product:", top_products.index[0])

    # Feature 3
    country_sales = df.groupby("Country")["TotalAmount"].sum().sort_values(ascending=False).head(10)
    print("Feature 3 - Top Country:", country_sales.index[0])

print("\n✅ All 10 test runs completed successfully.")
```
