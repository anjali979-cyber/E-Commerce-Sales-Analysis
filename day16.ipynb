Importing required Libraries

Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("data.csv",encoding="latin")
print(df.head())
```

Handeling the Missing Values

```python
df["Description"] = df["Description"].fillna("Unknown")
df["CustomerID"] = df["CustomerID"].fillna(0)

df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]
```

```python
print(df.info())
```

Implementing First Advanced Feature

```python
print("========== SALES SUMMARY DASHBOARD ==========")

total_sales = df["TotalAmount"].sum()
total_orders = df["InvoiceNo"].nunique()
total_products = df["Quantity"].sum()
average_order_value = total_sales / total_orders

print(f"Total Sales: {total_sales:.2f}")
print(f"Total Orders: {total_orders}")
print(f"Total Products Sold: {total_products}")
print(f"Average Order Value: {average_order_value:.2f}")
```

Implementing Second Advanced Feature

```python
print("===== Top 5 Customers by Total Purchase =====")

top_customers = (
    df.groupby("CustomerID")["TotalAmount"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

print(top_customers)
```

Connecting Both Features


```python
print("========== E-COMMERCE SALES ANALYSIS ==========")

# Advanced Feature 1
print("\n===== Sales Summary Dashboard =====")
print(f"Total Sales: {total_sales:.2f}")
print(f"Total Orders: {total_orders}")
print(f"Total Products Sold: {total_products}")
print(f"Average Order Value: {average_order_value:.2f}")

# Advanced Feature 2
print("\n===== Top 5 Customers by Total Purchase =====")
print(top_customers)
```

Verifying Both Features

```python
print("========== FULL PROJECT TEST ==========")

print("\nStep 1: Testing Sales Summary Dashboard...")
print(f"Total Sales: {total_sales:.2f}")
print(f"Total Orders: {total_orders}")
print(f"Total Products Sold: {total_products}")
print(f"Average Order Value: {average_order_value:.2f}")
print("✅ Dashboard Working")

print("\nStep 2: Testing Top 5 Customers...")
print(top_customers)
print("✅ Top 5 Customers Feature Working")

print("\n🎉 Full Project Test Completed Successfully!")
```
