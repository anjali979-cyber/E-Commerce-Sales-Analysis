Handeling Missing Values

```python
df["Description"] = df["Description"].fillna("Unknown")
```

```python
print(df["Description"].isnull().sum())
```

```python
df["CustomerID"] = df["CustomerID"].fillna("Unknown")
```

```python
print(df["CustomerID"].isnull().sum())
```

```python
print(df.columns)
```

Plot Distributions

```python
plt.figure(figsize=(8,5))
plt.hist(df["Quantity"], bins=30)
plt.title("Distribution of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["UnitPrice"], bins=30)
plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df["Country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Number of Orders")
plt.show()
```

Important Insights

## Key Insights

1. The Quantity distribution shows that most orders have small quantities, while a few orders have very large quantities.

2. The Unit Price distribution shows that most products have low prices, while only a few products are sold at high prices.

3. The Country analysis shows that sales are concentrated in a few countries, with the top countries contributing the highest number of orders.
