Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("data.csv",encoding="latin")
print(df.head())
```

Handeling the Missing Values

```python
df["Description"] = df["Description"].fillna("Unknown")
df["CustomerID"] = df["CustomerID"].fillna(0)

df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]
```

```python
print(df.info())
```

# 📊 E-Commerce Sales Analysis Dashboard

## Project Overview

This project analyzes e-commerce sales data and provides business insights through different analytical features.

### Features Included

- Sales Summary Dashboard
- Top 10 Best Selling Products
- Top 10 Countries by Total Sales
- Top 5 Customers by Total Purchase

**Project Status:** ✅ Working Successfully

```python
print("=" * 60)
print("      E-COMMERCE SALES ANALYSIS DASHBOARD")
print("=" * 60)
```

```python
total_sales = df["TotalAmount"].sum()
total_orders = df["InvoiceNo"].nunique()
total_products = df["Quantity"].sum()
average_order_value = total_sales / total_orders

print("\n📊 SALES SUMMARY DASHBOARD")
print("-" * 40)
print(f"Total Sales           : {total_sales:.2f}")
print(f"Total Orders          : {total_orders}")
print(f"Total Products Sold   : {total_products}")
print(f"Average Order Value   : {average_order_value:.2f}")
```

```python
top_products = (
    df.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print("\n🏆 TOP 10 BEST SELLING PRODUCTS")
print("-" * 40)
print(top_products)
```

```python
top_countries = (
    df.groupby("Country")["TotalAmount"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print("\n🌍 TOP 10 COUNTRIES BY TOTAL SALES")
print("-" * 40)
print(top_countries)
```

```python
top_customers = (
    df.groupby("CustomerID")["TotalAmount"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

print("\n🏆 TOP 5 CUSTOMERS")
print("-" * 40)
print(top_customers)
```
