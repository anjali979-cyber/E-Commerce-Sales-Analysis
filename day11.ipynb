Adding Input Validation

```python
# Function to calculate TotalAmount with input validation

def calculate_total_amount(quantity, unit_price):
    if not isinstance(quantity, (int, float)):
        return "Error: Quantity must be a number."

    if not isinstance(unit_price, (int, float)):
        return "Error: Unit Price must be a number."

    if quantity < 0:
        return "Error: Quantity cannot be negative."

    if unit_price < 0:
        return "Error: Unit Price cannot be negative."

    return quantity * unit_price
```

```python
print(calculate_total_amount(5, 100))
```

Adding error handeling (try/catch)

```python
def safe_calculate_total_amount(quantity, unit_price):
    try:
        if quantity < 0 or unit_price < 0:
            return "Error: Quantity and Unit Price must be non-negative."

        total = quantity * unit_price
        return total

    except TypeError:
        return "Error: Please enter numeric values only."

    except Exception as e:
        return f"Unexpected Error: {e}"
```

```python
print(safe_calculate_total_amount(10, 20))
```

Testing with invalid inputs: empty strings, wrong types, edge cases

```python
# Test 1: Wrong data type
print(safe_calculate_total_amount("five", 100))

# Test 2: Empty string
print(safe_calculate_total_amount("", 100))

# Test 3: Negative quantity
print(safe_calculate_total_amount(-5, 100))

# Test 4: Negative unit price
print(safe_calculate_total_amount(5, -100))

# Test 5: Valid input
print(safe_calculate_total_amount(5, 100))
```

Verifying error messages

```python
# Function with clear error messages

def safe_calculate_total_amount(quantity, unit_price):
    try:
        if quantity < 0:
            return "Error: Quantity cannot be negative. Please enter 0 or a positive number."

        if unit_price < 0:
            return "Error: Unit Price cannot be negative. Please enter 0 or a positive number."

        total = quantity * unit_price
        return f"Success! Total Amount = {total}"

    except TypeError:
        return "Error: Invalid input. Please enter numeric values for Quantity and Unit Price."

    except Exception as e:
        return f"Unexpected Error: {e}"
```

```python
print(safe_calculate_total_amount(5, 100))
print(safe_calculate_total_amount(-5, 100))
print(safe_calculate_total_amount("abc", 100))
```
