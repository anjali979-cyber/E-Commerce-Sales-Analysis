Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("data.csv",encoding="latin")
print(df.head())
```

Handeling the Missing Values

```python
df["Description"] = df["Description"].fillna("Unknown")
df["CustomerID"] = df["CustomerID"].fillna(0)

df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]
```

```python
print(df.info())
```

Checking If any Operations are slow

```python
import time
```

```python
start_time = time.time()

top_products = (
    df.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

end_time = time.time()

print("Top 10 Best Selling Products")
print(top_products)

print(f"\nExecution Time: {end_time - start_time:.6f} seconds")
```

# Performance Optimization

The execution time of the Top 10 Best Selling Products feature was measured.

The operation completed successfully and no significant slowdown was observed.

Reducing Memory

```python
print("Memory Usage Before Optimization:")
print(df.memory_usage(deep=True))

# Convert CustomerID to integer
df["CustomerID"] = df["CustomerID"].astype(int)

print("\nMemory Usage After Optimization:")
print(df.memory_usage(deep=True))

print("\n✅ Memory optimization completed.")
```

# Memory Optimization

The dataset memory usage was checked.

CustomerID was converted to integer type to reduce memory usage where possible.

The dataset continues to work correctly after optimization.
