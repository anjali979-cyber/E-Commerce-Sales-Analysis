Loading the Dataset

```python
import pandas as pd
df = pd.read_csv("data.csv",encoding="latin")
print(df.head())
```

Handeling the Missing Values

```python
df["Description"] = df["Description"].fillna("Unknown")
```

```python
df["CustomerID"] = df["CustomerID"].fillna("Unknown")
```

Implementing Second Core Feature

```python
# Feature 2: Top 10 Best Selling Products

top_products = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)

print("Top 10 Best Selling Products:")
print(top_products)
```

Again creating Feature core 1

```python
df["TotalAmount"] = df["Quantity"]*df["UnitPrice"]
```

```python
print(df.columns)
```

Connecting Feature 2 to Feature 1

```python
top_sales = (
    df.groupby("Description")[["TotalAmount"]]
    .sum()
    .sort_values(by="TotalAmount", ascending=False)
    .head(10)
)

print(top_sales)
```
